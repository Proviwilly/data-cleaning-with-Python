```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
df = pd.read_excel('Data Cleaning Start.xlsx',sheet_name=1,header=1)
df.head()
```

```python
df.shape
```

```python
print(df.info(), "\n")
print(df.dtypes)
```

```python
df=df.drop(columns='Unnamed: 0')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
null_percentage=(df.isnull().sum()/df.shape[0]*100)
print('The percentage of null values are:', null_percentage)
```

```python
df['Contact']=df.Contact.str.strip().str.title()
```

```python
df.head()
```

```python
df[['Department', 'Region']]=df.Department.str.split('_', expand=True)
```

```python
df.head()
```

```python
df[['Client', 'quat']]=df.Client.str.split('(', expand=True)
df.head()
```

```python
df=df.drop(columns='quat')
df.head()
```

```python
print(df.info(), "\n")
print(df.dtypes)
```

```python
null_percentage=(df.isnull().sum()/df.shape[0]*100)
print('The percentage of null values are:', null_percentage)
```

```python
df['Payment'].fillna(df['Payment'].mode())
df['Revenue'].fillna(df['Revenue'].median())
```

```python
df.describe(include='all')
```

```python
df['Revenue']=df['Revenue'].fillna(df['Revenue'].median())
```

```python
df[df.isna().any(axis=1)]
```

```python
df['Payment']=df['Payment'].fillna(df['Payment'].mode(dropna=True).iloc[0])
```

```python
df['Payment'].mode()
```

```python
null_percentage=(df.isnull().sum()/df.shape[0]*100)
print('The percentage of null values are:', null_percentage)
```
